```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
activity = pd.read_csv('ACTIVITY/data.csv')
activity
```

```python
nosteps = activity['steps']>100
steps = activity.loc[nosteps,'steps']
days = len(steps)
print(steps)
```

```python
plt.figure(figsize=(15,7))
plt.plot(np.arange(days),steps)
plt.grid(True)
plt.xlabel('Days')
plt.ylabel('Number of steps')
plt.title('Steps Analysis')
plt.savefig('steps.png')
```
